# Introduction to handling audio with python
This is sample code to read a .wav-file and plot its waveform and histogram. 


```python
%matplotlib inline
from pylab import*
from scipy.io import wavfile # alternatively use pysoundfile
import numpy as np
```

```python
samplerate, snd = wavfile.read('../../res/misc/440_sine.wav')
snd.dtype

length, num_bins = snd.shape

# calculate the duration 
print(length/samplerate, ' sec')

# convert to floating point (-1.0 - 1.0)
snd = snd / (2.**15)

# select channel 0
ch1= snd[:,0] 
```

```python
# Plot waveform
timeArray = arange(0, length, 1)#contains sample number (0,1,2,3,...)
timeArray = timeArray / samplerate #contains time label in seconds
timeArray = timeArray * 1000  #scale to milliseconds

plot(timeArray, ch1, color='k')
ylabel('Amplitude')
xlabel('Time (ms)')
```

```python
# Plot histogram
hist, bins = np.histogram(ch1,100,(-1.0,1.0))
width = 0.7 * (bins[1] - bins[0])

center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width)
plt.show()  
```
